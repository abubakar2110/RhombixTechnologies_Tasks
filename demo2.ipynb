```python
def recommend_song(recent_playlist, data):
    recent_songs = data[data['song_id'].isin(recent_playlist)]
    recent_features = recent_songs[numerical_features].mean().values.reshape(1, -1)
    
    similarities = cosine_similarity(recent_features, data[numerical_features])
    data['similarity'] = similarities[0]
    
    recommended_song = data[~data['song_id'].isin(recent_playlist)].sort_values(by='similarity', ascending=False).iloc[0]
    return recommended_song['song_id'], recommended_song['song_title'],recommended_song['URL']

recent_playlist = [1,3,5,6,2000,1000]  # Replace with actual song IDs from the user's recent playlist
recommended_song_id, recommended_song_name ,URL= recommend_song(recent_playlist, data)
print(f"Recommended Song: {recommended_song_name} (ID: {recommended_song_id},{URL})")
```
